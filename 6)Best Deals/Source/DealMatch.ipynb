```python
import re
from TwitterAPI import TwitterAPI, TwitterOAuth, TwitterRestPager
o = TwitterOAuth.read_file('credentials.txt')
twitter = TwitterAPI(o.consumer_key,
                 o.consumer_secret,
                 o.access_token_key,
                 o.access_token_secret)
screen_name = 'BestBuy_Deals'
timeline = [tweet for tweet in twitter.request('statuses/user_timeline',
                                                {'screen_name': screen_name,
                                                 'count': 200})]
print ('got %d tweets for user %s' % (len(timeline), screen_name))

```

```python
for tweet in timeline:
    deal = tweet['text']
    print (deal + '\n')
```

```python
import re
import pymysql

cnx = pymysql.connect(user='root', password='root',
                              host='127.0.0.1',
                              database='bestdeals')
cursor = cnx.cursor()

query = ("SELECT id,nm FROM products")
cursor.execute(query)
```

```python
dealMatchGauranteed={}
for product in cursor:
    for tweet in timeline:
        deal = (tweet['text'])
        if deal.find(str(product[1])) >= 1:
            dealMatchGauranteed[str(product[0])]=deal
```

```python
print (dealMatchGauranteed)
    
```

```python
dealMatchFile = open('DealMatches.txt', 'w')

for deal in dealMatchGauranteed.keys():
  dealMatchFile.write(deal+"<!split!>"+dealMatchGauranteed[deal]+'\n')

dealMatchFile.close()
            
```
